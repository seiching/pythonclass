```
!ls

from gtts import gTTS

from pydub import AudioSegment
foldername='/content/drive/My Drive/AIAcourse/aicoach/'
print(foldername)
txtfilename=input('請輸入字幕原稿檔名含副檔名，字幕建議用標點符號，產生字幕時會自動換行，如果要換頁請換行')
def speak(foldername,sentence,filename):
    tts=gTTS(text=sentence,lang="zh-TW",slow=False)
    tts.save(foldername+filename)
with open(foldername+txtfilename,'r') as caption:
  i=0
  for line in caption.readlines():
    if(line[0]=='0' or len(line)<2):
      pass
    else:
      i=i+1
      print(str(i)+'.mp3')
      filename=str(i)+'.mp3'
      print(filename)
      speak(foldername+'mp3/',line,filename)
      print(line.strip()) # 把末尾的'\n'刪掉
  print(i)
```
